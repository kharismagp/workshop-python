## 12. Virtual Environments and Packages

### 12.1. Introduction

Aplikasi python sering kali menggunakan packages dan modul yang tidak ada pada standar library. aplikasi sering kali membutuhkan versi tertentu dari library.

Ini artinya tidak mungkin untuk aplikasi python memenuhi setiap aplikasi. Solusi dari masalah ini adalah virtual environtment, aplikasi yang berbeda dapat menggunakan virtual environtment yang berbeda pula.

## 12.2. Creating Virtual Environments

Modul yang digunakan untuk mengatur virtual environtment adalah venv. venv biasanya akan menginstall versi python yang paling akhir yang tersedia. jika memunyai versi python yang berbeda pada maka dapat menggunakan perintah lebih spesifik misalnya python3 atau versi berapa yang ingin anda gunakan.

```python
!python3 -m venv tutorial-env
```

Perintah di atas akan membuat environtment baru dengan nama tutorial-env jika itu belum ada.

Pada Windows jalankan:

```python
tutorial-env\Scripts\activate.bat
```

Pada MacOS jalankan:

```python
!source tutorial-env/bin/activate
```

Perintah di atas digunakan untuk mengaktifkan environtment jika environtment tersebut sudah dibuat

### 12.3. Managing Packages with pip

kita bisa menginstall, mengupgrade atau menghapus packages menggunakan program bernama pip. kita bisa mencary python package seperti berikut :

```python
!pip search astronomy
```

pip mempunyai banyak subcommand: "search", "install", "uninstall", "freeze" dan sebagainya.

kita bisa menginstall versi terakhir dari sebuah packages dengan nama packagenya

```python
!pip install novas
```

Atau kita dapat menginstal versi spesifik suatu paket dengan memberikan nama paket diikuti dengan `==` dan nomor versi:

```python
!pip install requests==2.6.0
```

Jika kita menjalankan kembali script di atas maka `pip` akan mengetahui bahwa sudah terinstall dan tidak akan melakukan apa-apa. kita bisa mengubah ke versi lain dengan mengganti nomor versiatau menggunakan `pip install --upgrade`

```python
!pip install --upgrade requests
```

`pip uninstall` diikuti dengan satu atau beberapa nama package akan menghapus package dari virtual environtment. pip show akan menampilkan informasi tentang package tertentu

```python
!pip show matplotlib
```

```python
!pip list
```

`pip freeze` akan membuat list sejenis dengan package yang terinstall, tapi bisa membuat output dalam bentuk data dengan format tertentu

```python
!pip freeze > requirements.txt
```

```python
cat requirements.txt
```

`requirements.txt` dapat digunakan sebagai version control dan di masukkan sebagai bagian dari aplikasi. User dapat menginstall semua package yang dibutuhkan dengan `install -r`
